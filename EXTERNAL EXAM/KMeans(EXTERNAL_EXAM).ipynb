```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv('/content/wine-clustering.csv')
df.head()
```

```python
x = df[['Alcohol','Malic_Acid']]
y = df[['Proline']]
```

```python
print(x)
```

```python
print(y)
```

```python
plt.scatter(x['Alcohol'],x['Malic_Acid'],c=y.values.ravel())
```

```python
km=KMeans(n_clusters=2)
km.fit(x)
centers=km.cluster_centers_
labels=km.labels_
print(labels)
print(y.values.ravel())

```

```python
from IPython.core.pylabtools import figsize
from matplotlib import axes
fig,axes=plt.subplots(1,2,figsize=(16,4))
axes[0].scatter(x['Alcohol'],x['Malic_Acid'],c=y.values.ravel())
axes[1].scatter(x['Alcohol'],x['Malic_Acid'],c=labels)






```
